```python
import os
import time
import sympy
import subprocess
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

def run_one(show_detail,compare_content):
    # os.system("/Users/dfpmts/Desktop/JUCE_Demos/NewProject/build/Project1/Project1")
    # os.system('"D:/fa23/Athernet/build/Project1/Debug/ConsoleApp/Project1.exe"')
    # subprocess.Popen("D:/fa23/Athernet/build/Project1/Debug/ConsoleApp/Project1.exe",shell=True).wait()

    print("start processing")
    time.sleep(3)    

    f = open("preamble.txt")
    preamble = []
    for line in f:
        for i in line.split(' '):
            if i != '':
                preamble.append(float(i))

    preamble_len = len(preamble)

    f = open("carrier_0.txt")
    carrier_0 = []
    for line in f:
        for i in line.split(' '):
            if i != '':
                carrier_0.append(float(i))
    f.close() 

    f = open("carrier_1.txt")
    carrier_1 = []
    for line in f:
        for i in line.split(' '):
            if i != '':
                carrier_1.append(float(i)) 
    f.close() 

    sent = []
    f = open("sent.txt")
    for line in f:
        for i in line.split(' '):
            if i != '':
                sent.append(int(i))
    f.close()         

    f = open("received.txt")
    a = []
    for line in f:
        for i in line.split(' '):
            if i != '':
                a.append(float(i))
    f.close() 

    # f = open("packet_length.txt")
    # for line in f:
    #     for i in line.split(' '):
    #         if i != '':
    #             packet_len = int(i)

    packet_len = 10

    f.close() 
    if show_detail:
        print("Packet Length:  ",str(packet_len))
    # x = np.correlate(a,preamble,'same')

    mpl.rcParams['agg.path.chunksize'] = 10000

    if show_detail:    
        plt.figure(figsize=(100,50))        
        plt.plot(a)
        # plt.ylim(0,0.3)
        plt.show()

    # plt.plot(x)
    percent = []

    points = []

    max = 0
    pos = 0

    running_average = 0

    max_avg = 0

    for i in range(0,len(a)-preamble_len):
        res = np.dot(a[i:i+preamble_len],preamble)

        res_percent = res*res/(np.dot(preamble,preamble)*np.dot(a[i:i+preamble_len],a[i:i+preamble_len]))

        running_average += (res_percent-running_average)/(i+1)
        if running_average > max_avg:
            max_avg = running_average
        percent.append(res_percent)

        if((res_percent > running_average*3 and res_percent>0.2) and (res > max)):
            max = res
            pos = i
        if i > pos + 300 and max!=0:
            points.append(pos)
            max=0
            pos=0;
    if show_detail:    
        print(max_avg)
        plt.figure(figsize=(100,50))        
        plt.plot(percent)
        plt.ylim(0,0.3)
        plt.show()

    total_data = []
    total_length = 0

    index = 0

    crc8 = [1,1,1,0,1,0,1,0,1]
    crc_total_error = 0
    for pos in points:
        index+=1
        # print(index)
        if show_detail:
            print(pos)
        piece = len(carrier_0)
        # print(piece)
        pos+=preamble_len
        res_0 = []
        res_1 = []
        data = []
        for i in range(pos,pos+8*piece+packet_len*piece+8*piece+200*piece,piece):
            dot_0 = np.dot(a[i:i+piece],carrier_0)
            dot_1 =np.dot(a[i:i+piece],carrier_1)
            
            if dot_0 > 0:
                data.append(0)
            else:
                data.append(1)
            res_0.append(dot_0)
            res_1.append(dot_1)

        # if show_detail:
        #     plt.figure(figsize=(100,50))
        #     plt.plot(res_0[100:])
        #     plt.plot(res_1[100:])
        #     plt.show()

        length = 0
        for i in range(0,8):
            if data[i]:
                length+=2**i
        total_length += length;            
        
        if show_detail:
            print("Length "+str(index)+": "+str(length))        
            print(data[8:8+packet_len])
        total_data += data[8:8+packet_len]

        data_to_check = data.copy()
        for i in range(0,(8+packet_len)):
            if data_to_check[i] == 1:
                for j in range(0,9):
                    data_to_check[i+j] ^= crc8[j]
        error = False
        for i in range(8+packet_len,8+packet_len+8):
            if data_to_check[i] != 0:
                error = True;
        if error:
            crc_total_error+=1
            if show_detail:
                print(" !!! Error CRC check FAILED  !!! ")
            for i in range(8+packet_len,8+packet_len+8):
                if show_detail:
                    print(data_to_check[i],end='')
        if show_detail:                            
            print()

    # print(sent)
    # print(total_data)
    if show_detail:
        print("------ Length: "+str(total_length))
    if compare_content:
        total_errors = 0

        for i in range(0,len(sent),packet_len):
            errors=0
            for j in range(0,packet_len):
                if sent[i+j] != total_data[i+j]:
                    # print("-----!!!!!!!-----   "+str(i));
                    errors+=1
            if show_detail:            
                print(str(int(i/packet_len+1))+" ------- ", errors);
            total_errors+=errors

        if show_detail:
            print("------ Errors:     ",str(total_errors))

    if show_detail:
        print("------ Bad Frames: ",str(crc_total_error))

        # plt.plot(data)
        # plt.show()
        # plt.plot(a)
        # plt.show()
        # plt.plot(x)
        # plt.show()
    total_packets_sent = int(len(sent)/packet_len)
    lost_packets = len(points) - total_packets_sent
    return (crc_total_error+lost_packets)/total_packets_sent
```

```python
from IPython.display import clear_output
now_id = 0
avg = 0
while True:
    erasure_rate = run_one(show_detail=False,compare_content=False)
    avg += (erasure_rate-avg)/(now_id+1)
    now_id+=1
    clear_output(wait=True)
    print("    Now:  ",erasure_rate)
    print("Average:  ", avg)
```
